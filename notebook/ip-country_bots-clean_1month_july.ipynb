```python
def agg_day(day):	bots = bots_arr[day-1][['COUNTRY', 'CNT']]	bots_totals = bots.groupby(['COUNTRY'], sort=True).sum()	bots_totals.columns = ['cnt_bots']	clean = clean_arr[day-1][['COUNTRY', 'CNT']]	clean_totals = clean.groupby(['COUNTRY'], sort=True).sum()	clean_totals.columns = ['cnt_clean']	dayagg = bots_totals.join(clean_totals)        dayagg = dayagg.fillna(0)	dayagg['perc_bots'] =  dayagg.cnt_bots / (dayagg.cnt_bots + dayagg.cnt_clean) * 100        dayagg['cnt_both'] = dayagg.cnt_bots + dayagg.cnt_clean	dayagg_s = dayagg.sort(columns='cnt_both', ascending=False)        return dayagg_s
```
